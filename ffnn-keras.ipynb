# Feedforward Neural Network built on Keras

Below is a program built on Keras to build a regression model. I benefit a lot from this link: https://machinelearningmastery.com/regression-tutorial-keras-deep-learning-library-python/

```python
import numpy as np
import pandas
import keras.callbacks
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
dataframe = pandas.read_csv("layton_wildprocessed", header = 0)
# print (dataframe)
```

```python
dataset = dataframe.values
dataset_new = np.delete(dataset, [1, 5], axis = 1)  #delete the second the sixth items (timestamps) in each array
#print (dataset_new)
X = dataset_new[:-1]
Y = dataset_new[1:]
Y = np.delete(Y, [0], axis = 1)
# print (X, Y)
```

```python
def baseline_model():
    model = Sequential()
    model.add(Dense(4, input_dim = 4, kernel_initializer = 'normal', activation = 'relu'))
    model.add(Dense(3, kernel_initializer = 'normal'))
    model.compile(loss = 'mean_squared_error', optimizer = 'adam')
    return model
```

```python
seed = 7
np.random.seed(seed)

estimator= KerasRegressor(build_fn=baseline_model, nb_epoch = 100, batch_size = 5, verbose = 2)
```

```python
kfold = KFold(n_splits = 10, random_state = seed)
results = cross_val_score(estimator, X, Y, cv= kfold)

#print (callback)
#print (estimator)
#print (results)
print ("Results: %.5f (%.5f) MSE" % (results.mean(), results.std()))
```

```python
# evaluate model with standardized dataset
np.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=baseline_model, epochs=50, batch_size=5, verbose=2)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.5f (%.5f) MSE" % (results.mean(), results.std()))


# print(history.losses)
# outputs
```
